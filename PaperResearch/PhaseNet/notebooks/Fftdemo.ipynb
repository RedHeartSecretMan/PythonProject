```python
import numpy as np
import matplotlib.pyplot as plt

# 产生正弦信号
Ts=0.001
t = Ts*np.array(range(10000))
x = np.sin(2*np.pi*250*t)
display(t)
display(x)

plt.figure(1)
plt.plot(t,x)
plt.show()
 
# FFT变换
X=np.fft.fft(x,np.size(x,0),axis=0)/x.size*2
freq=np.fft.fftfreq(np.size(x,0),Ts)

display(X)
display(np.real(X))
plt.figure(2)
plt.plot(freq,np.real(X))
plt.show()
```
